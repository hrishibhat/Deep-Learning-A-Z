```python
# AutoEncoders

# Importing the libraries
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.parallel
import torch.optim as optim
import torch.utils.data
from torch.autograd import Variable
```

```python
# Preparing the training set and the test set
training_set = pd.read_csv('lotto/lotto_6_balls_train.csv', header=None)
# training_set = np.array(training_set, dtype = 'int')
test_set = pd.read_csv('lotto/lotto_6_balls_test.csv', header=None)
# test_set = np.array(test_set, dtype = 'int')
```

```python
training_set = training_set.values.tolist()
test_set = test_set.values.tolist()


# training_set = training_set.values.T.tolist()
# test_set = test_set.values.T.tolist()


# training_set = list(training_set.values.T.flatten())
# test_set = list(test_set.values.T.flatten())
```

```python
# Converting the data into Torch tensors
training_set = torch.FloatTensor(training_set)
test_set = torch.FloatTensor(test_set)
```

```python
# Creating the architecture of the Neural Network
class SAE(nn.Module):
    def __init__(self, ):
        super(SAE, self).__init__()
        self.fc1 = nn.Linear(len(training_set[0]), 20)
        self.fc2 = nn.Linear(20, 10)
        self.fc3 = nn.Linear(10, 20)
        self.fc4 = nn.Linear(20, len(training_set[0]))
        self.activation = nn.Sigmoid()
    def forward(self, x):
        x = self.activation(self.fc1(x))
        x = self.activation(self.fc2(x))
        x = self.activation(self.fc3(x))
        x = self.fc4(x)
        return x
sae = SAE()
criterion = nn.MSELoss()
optimizer = optim.RMSprop(sae.parameters(), lr = 0.01, weight_decay = 0.5)

```

```python
# Training the SAE
nb_epoch = 10
for epoch in range(1, nb_epoch + 1):
    train_loss = 0
    s = 0.
    for lotto_result in range(len(training_set)):
        input = Variable(training_set[lotto_result]).unsqueeze(0)
        target = input.clone()
        print ('target:', target.data)
        if torch.sum(target.data > 0) > 0:
            output = sae(input)
            target.require_grad = False
            output[target == 0] = 0
            print ('output:', torch.round(output.data))
            loss = criterion(output, target)
            accuracy = torch.eq(target.data, torch.round(output.data))
            print ("Total Match:", torch.sum(accuracy))
            print ("Accuracy:", torch.sum(accuracy)/len(training_set[0]))
            mean_corrector = len(training_set[0])/float(torch.sum(target.data > 0) + 1e-10)
            loss.backward()
            train_loss += np.sqrt(loss.data[0]*mean_corrector)
            s += 1.
            optimizer.step()
    print('epoch: '+str(epoch)+' loss: '+str(train_loss/s))
```

```python
# Let's save the good model, as it was able to predict 4 balls 1 time
torch.save(sae.state_dict(), './sae1.pth')
```

```python
# Now, let' load the pre save model
sae.load_state_dict(torch.load('./sae.pth'))

# Testing the SAE
test_loss = 0
s = 0.
for lotto_result in range(len(test_set)):
        input = Variable(training_set[lotto_result]).unsqueeze(0)
        target = Variable(test_set[lotto_result])
        print ('target:', target.data)
        if torch.sum(target.data > 0) > 0:
            output = sae(input)
            target.require_grad = False
            output[target == 0] = 0
            print ('output:', torch.round(output.data))
            loss = criterion(output, target)
            accuracy = torch.eq(target.data, torch.round(output.data))
            print ("Total Match:", torch.sum(accuracy))
            print ("Accuracy:", torch.sum(accuracy)/len(training_set[0]))
            mean_corrector = len(test_set[3])/float(torch.sum(target.data > 0) + 1e-10)
            test_loss += np.sqrt(loss.data[0]*mean_corrector)
            s += 1.
print('test loss: '+str(test_loss/s))
```
